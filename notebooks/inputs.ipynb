```python
import pandas as pd
import mysql.connector

# Cargar los datos del archivo Excel ya limpio, omitiendo las primeras filas innecesarias
df = pd.read_excel('INPUT/INGRESO_SEP.xlsx', skiprows=7)

# Eliminar la columna 'Nº'
df = df.drop('Nº', axis=1)

# Seleccionar las columnas relevantes del DataFrame original
selected_columns = [
    'Código Ingreso', 'Fecha Ingreso', 'Motivo de Ingreso', 'Gerencia Solicitante',
    'Departamento Solicitante', 'Ubicacion', 'Tipo de Gasto', 'Partida Presupuestaria',
    'Proveedor', 'Código Item', 'Item', 'Unidad Medida', 'Grupo', 'Subgrupo',
    'Cantidad', 'Precio Unitario', 'Importe (Histórico) 100%', 'CF-IVA',
    'Importe Calculado'
]

# Crear un nuevo DataFrame con las columnas seleccionadas
df_selected = df[selected_columns]

# Definir los nuevos nombres en el orden deseado
new_column_names = [
    'codigo_ingreso', 'fecha_ingreso', 'motivo_ingreso',
    'gerencia_solicitante', 'departamento_solicitante',
    'ubicacion', 'tipo_de_gasto', 'partida_presupuestaria', 'proveedor',
    'codigo_item', 'descripcion', 'unidad_medida',
    'grupo', 'sub_grupo', 'cantidad',
    'precio_unitario', 'importe_historico', 'cf_iva',
    'importe_calculado'
]

# Cambiar los nombres de las columnas
df_selected.columns = new_column_names

# Agregar la nueva columna 'fecha' con los mismos datos de 'fecha_ingreso'
df_selected['fecha'] = pd.to_datetime(df_selected['fecha_ingreso'], dayfirst=True).dt.strftime('%Y-%m-%d')

# Copiar los datos de 'codigo_item' a una nueva columna 'codigo_partida'
df_selected['codigo_partida'] = df_selected['codigo_item']

# Conectar a la base de datos
connection = mysql.connector.connect(
    host='192.168.100.50',        # Cambia por tu host
    user='mantto',                # Cambia por tu usuario
    password='Sistemas0,',        # Cambia por tu contraseña
    database='Catalogo'           # Cambia por tu base de datos
)

# Crear un cursor para ejecutar consultas
cursor = connection.cursor()

# Definir la consulta SQL para insertar los datos
insert_query = """
INSERT INTO ingreso_copia_rmc (
    codigo_ingreso, fecha_ingreso, motivo_ingreso, gerencia_solicitante, 
    departamento_solicitante, ubicacion, tipo_gasto, partida_presupuestaria, 
    codigo_partida, proveedor, codigo_item, descripcion, unidad_medida, grupo, 
    sub_grupo, cantidad, precio_unitario, costo_historico, cf_iva, importe_calculado, fecha
) 
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

# Iterar sobre las filas del DataFrame e insertar cada una en la base de datos
for index, row in df_selected.iterrows():
    try:
        # Ejecutar la consulta de inserción
        cursor.execute(insert_query, tuple(row))
    except mysql.connector.Error as err:
        print(f"Error en la fila {index}: {err}")
        continue  # Continuar con la siguiente fila si ocurre un error

# Confirmar la transacción
connection.commit()

# Cerrar el cursor y la conexión
cursor.close()
connection.close()

print("Datos insertados exitosamente.")
```
